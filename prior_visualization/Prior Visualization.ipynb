### Construct samples from prior distributions and save

```python
# Mean for prior if using normal distribution
strike = 205.0
length = 270000.
width = 80000.
depth = 20.0
slip = 9.
rake = 90.
dip = 15.
longitude = 132.4
latitude = -6.485
means = np.array([strike, length, width, depth, slip, rake, dip,
    longitude, latitude])

# Standard deviations
strike_std = 10.
length_std = 33000.
width_std = 15000.
depth_std = 3.3
slip_std = 2.
rake_std = 3.
dip_std = 1.6
longitude_std = .27
latitude_std = .6
stds = np.array([strike_std, length_std, width_std, depth_std,
    slip_std, rake_std, dip_std, longitude_std, latitude_std])
```

```python
# Draw 200,000 from prior and save so in good format for make_plots.py
num_draws = 200000
data = []
for mean, std in zip(means, stds):
    prior = stats.norm(mean, std)
    data.append(prior.rvs(size=num_draws))

data = np.vstack((np.zeros(11), np.hstack((np.array(data).T, 2*np.ones((num_draws, 2))))))
np.save("prior.npy", data)
```

## 9X9 Matrix

```python
def make_matrix(prior=False):
    plt.rcParams['figure.figsize'] = [16,16]
    params = ['strike', 'length', 'width', 'depth', 'slip', 'rake', 'dip', 'longitude', 'latitude']
    for i, param2 in enumerate(params):
        for j, param1 in enumerate(params):
            plt.subplot(9, 9, i*9+j+1)
            title = ""
            if i == 0:
                title = param1
            if param1 == param2:
                param = param1
                xlower, xupper = bounds[param]
                make_hist(param, prior=prior, title=title, matrix=True)
                plt.xlim(xlower, xupper)
                plt.xticks([])
                plt.yticks([])
                if j == 0:
                    plt.ylabel(param2)
            else:
                bins = 30
                xlower, xupper = bounds[param1]
                ylower, yupper = bounds[param2]
                make_2dhist(param1, param2, bins, prior=prior, colorbar=False, title=title, matrix = True)
                plt.xlim(xlower, xupper)
                plt.ylim(ylower, yupper)
                plt.xticks([])
                plt.yticks([])
                if j == 0:
                    plt.ylabel(param2)

    plt.show()
```
